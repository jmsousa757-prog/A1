```python


print("==============================================")
print("     Sistema de Entrada   ")
print("==============================================")

# 1. Tentar pegar a idade e j√° converter
try:
    # input() sempre devolve uma string (texto)
    entrada = input("Qual a sua idade? ")

    # int() tenta converter a string para um n√∫mero inteiro
    # Se o usu√°rio digitar "abc", isso vai falhar e pular pro 'except'
    idade = int(entrada)

# 2. O "Guarda" (Guard Clause)
    # Primeiro, vamos barrar as idades sem sentido (negativas ou muito altas).
    # Se a idade for inv√°lida, a gente j√° para o c√≥digo aqui.
    if idade <= 0 or idade > 120:
        print(f"\n[IDADE INEXISTENTE] ü§®")
        print(f"Idade '{idade}'? Isso n√£o √© v√°lido. Tente de novo.")

# 3. L√≥gica Principal (O "Caminho Feliz")
    # Se o c√≥digo chegou at√© aqui, a idade √© um n√∫mero v√°lido (ex: 15, 20, 50).

    # Regra 1: Maior de idade
    elif idade >= 18:
        print(f"\n[APROVADO] ‚úÖ")
        print(f"Idade: {idade}. Acesso total. Pode pegar a pulseira VIP.")

    # Regra 2: 16 ou 17 anos
    # S√≥ entra aqui se o 'if' de cima (>= 18) for falso.
    elif idade >= 16:
        print(f"\n[APROVADO COM RESTRI√á√ÉO] ‚ö†Ô∏è")
        print(f"Idade: {idade}. Acesso √† pista, mas longe do bar. Pulseira Jovem.")

    # Regra 3: Menor de 16
    # Se nenhum dos 'if' ou 'elif' acima foi verdadeiro, cai aqui.
    else:
        print(f"\n[BARRADO] ‚õî")
        print(f"Idade: {idade}. Desculpe, festa s√≥ para maiores de 16.")

# 4. Tratamento de Erro (O Famoso "try...except")
# Se o 'try' l√° em cima falhar (ex: digitar "vinte"), o Python pula direto pra c√°.
except ValueError:
    print(f"\n[ERRO DE ENTRADA] ‚ùå")
    print(f"O que voc√™ digitou ('{entrada}') n√£o √© um N√öMERO.")
    print("Rode o programa de novo e digite sua idade com n√∫meros.")

print("\n==============================================")
```
